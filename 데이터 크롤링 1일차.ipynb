#🟦 Command Mode 단축키 (셀 조작용)

    단축키 (맥 기준)	기능
    A	위에 셀 추가
    B	아래에 셀 추가
    X	셀 잘라내기
    C	셀 복사
    V	아래에 붙여넣기
    Shift + V	위에 붙여넣기
    D, D (빠르게 두 번)	셀 삭제
    Z	마지막 셀 복구 (Undo)
    M	셀을 Markdown으로 변경
    Y	셀을 코드(Code)로 변경
    ↑ / ↓	셀 위/아래 선택
    Shift + ↑ / ↓	여러 셀 선택
    Shift + M	선택한 셀 합치기
    Cmd + S	저장

#🟩 Edit Mode 단축키 (코딩용)

    단축키 (맥 기준)	기능
    Cmd + Enter	실행 (셀 실행만)
    Shift + Enter	실행 후 다음 셀로 이동 (없으면 새 셀 생성)
    Option + Enter	실행 후 아래에 새 셀 추가
    Tab	자동완성
    Shift + Tab	함수 설명 툴팁 보기
    Cmd + /	현재 줄 주석 토글
    Esc	Command Mode로 전환

#🔍 기타 유용한 팁

    모든 단축키 보기: H (Command Mode에서)
    셀 나누기: Cmd + Shift + - (Edit Mode에서 커서 기준 셀 나눔)
    커널 재시작: 0, 0 (Command Mode에서 0 두 번 누르기)

```python
import requests
from bs4 import BeautifulSoup

# 크롤링할 페이지 URL
url = 'http://books.toscrape.com/catalogue/page-1.html'

# 페이지 요청
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# 책 목록을 포함하는 모든 태그 선택
books = soup.select('article.product_pod')

for book in books:
    # 제목은 태그 속 title 속성에 들어 있음
    title = book.h3.a['title']

    # 가격은 클래스 'price_color' 사용
    price = book.select_one('.price_color').text

    # 평점은 클래스 이름에 포함되어 있음 (예: star-rating Three)
    rating_class = book.select_one('.star-rating')['class']
    # 두 번째 요소가 평점 이름이야 (e.g. 'Three' → 3)
    rating_str = rating_class[1]

    # 평점을 숫자로 변환
    rating_dict = {
        'One': 1,
        'Two': 2,
        'Three': 3,
        'Four': 4,
        'Five': 5
    }
    rating = rating_dict.get(rating_str, 0)

    print(f'제목: {title}\n가격: {price}\n평점: {rating}점\n---')
```

```python
import requests
from bs4 import BeautifulSoup

book_data = []  # 여기에서 데이터를 담을 빈 리스트를 먼저 선언!

rating_dict = {
    'One': 1,
    'Two': 2,
    'Three': 3,
    'Four': 4,
    'Five': 5
}

for page in range(1, 6):
    url = f'http://books.toscrape.com/catalogue/page-{page}.html'
    res = requests.get(url)

    res.encoding = 'utf-8'  # ★ 인코딩 명시 추가 ★

    soup = BeautifulSoup(res.text, 'html.parser')
    
    books = soup.select('article.product_pod')
    
    for book in books:
        title = book.h3.a['title']
        price = book.select_one('.price_color').text

        # 잘못된 문자 제거 (Â 등)
        price = price.replace('Â', '').replace('£', '').strip()
        price = float(price)

        rating_class = book.select_one('.star-rating')['class']
        rating_str = rating_class[1]
        rating = rating_dict.get(rating_str, 0)
        
        book_data.append({
            '제목': title,
            '가격(£)': price,
            '평점': rating
        })
```

```python
import pandas as pd

df = pd.DataFrame(book_data)
df.head()  # 앞부분만 미리 보기
```

```python
# 평점 내림차순, 가격 오름차순으로 정렬
df_sorted = df.sort_values(by=['평점', '가격(£)'], ascending=[False, True])

# 추천 Top 5
recommendations = df_sorted
print("📚 추천 도서 목록 (Top 5)\n")
print(recommendations)
```
