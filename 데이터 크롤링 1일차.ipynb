```python
import requests
from bs4 import BeautifulSoup

book_data = []  # ì—¬ê¸°ì—ì„œ ë°ì´í„°ë¥¼ ë‹´ì„ ë¹ˆ ë¦¬ìŠ¤íŠ¸ë¥¼ ë¨¼ì € ì„ ì–¸!

rating_dict = {
    'One': 1,
    'Two': 2,
    'Three': 3,
    'Four': 4,
    'Five': 5
}

for page in range(1, 6):
    url = f'http://books.toscrape.com/catalogue/page-{page}.html'
    res = requests.get(url)

    res.encoding = 'utf-8'  # â˜… ì¸ì½”ë”© ëª…ì‹œ ì¶”ê°€ â˜…

    soup = BeautifulSoup(res.text, 'html.parser')
    
    books = soup.select('article.product_pod')
    
    for book in books:
        title = book.h3.a['title']
        price = book.select_one('.price_color').text

        # ì˜ëª»ëœ ë¬¸ì ì œê±° (Ã‚ ë“±)
        price = price.replace('Ã‚', '').replace('Â£', '').strip()
        price = float(price)

        rating_class = book.select_one('.star-rating')['class']
        rating_str = rating_class[1]
        rating = rating_dict.get(rating_str, 0)
        
        book_data.append({
            'ì œëª©': title,
            'ê°€ê²©(Â£)': price,
            'í‰ì ': rating
        })
```

```python
import pandas as pd

df = pd.DataFrame(book_data)
df.head()  # ì•ë¶€ë¶„ë§Œ ë¯¸ë¦¬ ë³´ê¸°
```

```python
# í‰ì  ë‚´ë¦¼ì°¨ìˆœ, ê°€ê²© ì˜¤ë¦„ì°¨ìˆœìœ¼ë¡œ ì •ë ¬
df_sorted = df.sort_values(by=['í‰ì ', 'ê°€ê²©(Â£)'], ascending=[False, True])

# ì¶”ì²œ Top 5
recommendations = df_sorted
print("ğŸ“š ì¶”ì²œ ë„ì„œ ëª©ë¡ (Top 5)\n")
print(recommendations)
```
